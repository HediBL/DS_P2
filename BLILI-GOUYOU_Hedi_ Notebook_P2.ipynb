```python
# Sélectionnez uniquement les colonnes des années 2005 à 2015
colonnes_selectionnees = ['Country Name', 'Region', 'GDP', 'Percentage of female population age', '% Internet user'] + [str(annee) for annee in range(2005, 2016)]
df_merged_selection = df_merged.filter(items=colonnes_selectionnees)
print(df_merged_selection.head())

```

```python
from sklearn.preprocessing import StandardScaler

# Créez une instance de StandardScaler
scaler = StandardScaler()

# Sélectionnez uniquement les colonnes numériques à normaliser (années 2005 à 2015)
colonnes_numeriques = [str(annee) for annee in range(2005, 2016)]
df_merged_numerique = df_merged_selection[colonnes_numeriques]

# Appliquez la normalisation sur les colonnes numériques
df_merged_normalise = scaler.fit_transform(df_merged_numerique)

# Remplacez les valeurs normalisées dans le DataFrame d'origine
df_merged_selection[colonnes_numeriques] = df_merged_normalise

# Affichez le DataFrame normalisé
print(df_merged_selection.head())
```

Moyenne de chaque indicateur par pays

```python
# Groupement par pays et calcul de la moyenne pour chaque indicateur
df_mean_by_country = df_merged_selection.groupby(['Country Name', 'Region']).mean().reset_index()

# Affichage du DataFrame résultant
print(df_mean_by_country)
```

```python
import matplotlib.pyplot as plt

# Sélectionnez les 5 premiers pays
top_5_countries = df_mean_by_country.head(5)

# Sélectionnez les colonnes d'indicateurs (à l'exception des colonnes 'Country Name' et 'Region')
indicator_columns = top_5_countries.columns[2:]

# Créez un graphique à barres pour chaque indicateur
plt.figure(figsize=(12, 8))
for indicator in indicator_columns:
    plt.bar(top_5_countries['Country Name'], top_5_countries[indicator], label=indicator)

# Ajoutez des étiquettes et une légende
plt.title('Moyenne de chaque indicateur pour les 5 premiers pays')
plt.xlabel('Pays')
plt.ylabel('Moyenne')
plt.xticks(rotation=45)
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))

# Affichez le graphique
plt.tight_layout()
plt.show()
```

```python
#Ce code génère un dendrogramme basé sur la matrice de distance entre les indicateurs des 5 premiers pays. 
#Chaque branche du dendrogramme représente un pays, et la hauteur des branches indique la distance entre les pays en termes de similarité des indicateurs. 
#Les pays qui sont regroupés à une faible distance sont similaires en termes de moyennes d'indicateurs

import scipy.cluster.hierarchy as sch

# Sélectionnez les 5 premiers pays
top_5_countries = df_mean_by_country.head(5)

# Sélectionnez les colonnes d'indicateurs (à l'exception des colonnes 'Country Name' et 'Region')
indicator_columns = top_5_countries.columns[2:]

# Créez une matrice de données pour les indicateurs des 5 premiers pays
data_matrix = top_5_countries[indicator_columns].values

# Calculez la matrice de distance
dendrogram = sch.dendrogram(sch.linkage(data_matrix, method='ward'),
                            labels=top_5_countries['Country Name'].tolist())  # Utilisez les noms des pays comme étiquettes

# Ajoutez des étiquettes aux axes
plt.title('Dendrogramme des indicateurs pour les 5 premiers pays')
plt.xlabel('Pays')
plt.ylabel('Distance')

# Affichez le dendrogramme
plt.xticks(rotation=45)  # Faites pivoter les étiquettes pour une meilleure lisibilité
plt.show()
```

```python
# Vérifiez les colonnes non numériques
non_numeric_columns = df_merged_selection.columns[~df_merged_selection.apply(pd.to_numeric, errors='coerce').notna().all()]
print(non_numeric_columns)
```

```python
# Exclure les colonnes 'Country Name' et 'Region' du calcul de la moyenne
numeric_columns = df_merged_selection.columns[2:]  # Sélectionnez les colonnes numériques
df_merged_selection.loc[:, 'Score'] = df_merged_selection[numeric_columns].mean(axis=1)  # Calculer la moyenne
# Afficher le DataFrame avec la colonne 'Score'
print(df_merged_selection[['Country Name', 'Score']])

```

```python

# Sélectionner les 20 premiers pays
top_20_countries = df_merged_selection.nlargest(20, 'Score')

# Créer un graphique à barres pour afficher le score par pays en utilisant Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Country Name', y='Score', data=top_20_countries)
plt.xlabel('Pays')
plt.ylabel('Score')
plt.title('Score des 20 premiers pays')
plt.xticks(rotation=90)
plt.show()
```

```python
# Sélectionner les 50 premiers pays
top_50_countries = df_merged_selection.nlargest(50, 'Score')

# Créer un graphique à barres pour afficher le score par pays en utilisant Seaborn
plt.figure(figsize=(15, 8))
sns.barplot(x='Country Name', y='Score', data=top_50_countries)
plt.xlabel('Pays')
plt.ylabel('Score')
plt.title('Score des 50 premiers pays')
plt.xticks(rotation=90)
plt.show()
```

Visualisation de la qualité de données au sein de la colonne YEAR

```python

# Calculer le pourcentage de valeurs nulles pour chaque année dans la colonne 'Year'
null_percentage_by_year = merged_df.groupby('Year').apply(lambda x: x.isnull().mean() * 100)

# Calculer la moyenne du pourcentage de valeurs nulles pour chaque année
mean_null_percentage_by_year = null_percentage_by_year.mean(axis=1).sort_values(ascending=False)

# Créer un graphique à barres pour visualiser les années avec le pourcentage de valeurs nulles le plus élevé
plt.figure(figsize=(8, 6))
mean_null_percentage_by_year.plot(kind='bar', color='skyblue')
plt.title('Pourcentage moyen de valeurs nulles par année')
plt.xlabel('Année')
plt.ylabel('Pourcentage moyen de valeurs nulles')
plt.tight_layout()

# Afficher le graphique
plt.show()
```

Visualisation de la qualité des données par pays

```python

# Calculer le pourcentage de valeurs nulles pour chaque pays dans la colonne 'Country Name'
null_percentage_by_country = merged_df.groupby('Country Name').apply(lambda x: x.isnull().mean() * 100)

# Calculer la moyenne du pourcentage de valeurs nulles pour chaque pays
mean_null_percentage_by_country = null_percentage_by_country.mean(axis=1).sort_values(ascending=False)

# Créer un graphique à barres horizontales pour visualiser les pays avec le pourcentage de valeurs nulles le plus élevé
plt.figure(figsize=(10, 12))
mean_null_percentage_by_country.plot(kind='barh', color='skyblue')
plt.title('Pourcentage moyen de valeurs nulles par pays')
plt.xlabel('Pourcentage moyen de valeurs nulles')
plt.ylabel('Pays')
plt.tight_layout()

# Afficher le graphique
plt.show()
```
